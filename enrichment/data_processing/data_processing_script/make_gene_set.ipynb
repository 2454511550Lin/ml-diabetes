```python
def get_gene_data(module_dict):
  gene_data={}
  for tissue in tissues:
    universal_genes=[]
    for sign in signs:
        for module in module_dict[(tissue,sign)].keys():
            genes = module_dict[(tissue,sign)][module]
            universal_genes.extend(genes)
    gene_data[tissue]=set(universal_genes)
  return gene_data
```

```python
tissues = ['islet', 'liver', 'adipose', 'kidney', 'gastroc']
signs = ['SIGNED', 'UNSIGNED']

# run different module using python process
path = 'module_dict.pickle'
with open(path, 'rb') as f:
        module_dict = pickle.load(f)

gene_data=get_gene_data(module_dict)
```

```python
for tissue in tissues:
    for sign in signs:
      directory=tissue+'/'
      if not os.path.exists(directory):
        os.makedirs(directory)
```

```python
# opening the file in read mode
my_file = open("discard_genes.txt", "r")
# reading the file
data = my_file.read()
discard_genes = set(data.split("\n"))
print(discard_genes)
my_file.close()
```

```python
def get_gene_symbol(gene):
  gene_symbol=''
  if gene.count('.')==2:
    if gene not in discard_genes:
      gene_symbol=gene.split('.',1)[1]
  else:
    gene_symbol=gene.split('.')[1]
  return gene_symbol
```

```python
## Saving data for modules in one file
for tissue in tissues:
    for sign in signs:
      directory=tissue+'/'
      df = pd.DataFrame(columns=["Accession","module","tissue_origin","kME_module","multiple_tissues","gene_symbol","entrez_ID","numb_tissues","Sign"])
      file_path=directory+sign+".csv"
      for module in module_dict[(tissue,sign)].keys():
        genes = module_dict[(tissue,sign)][module]
        for gene in genes:
          gene_symbol=get_gene_symbol(gene)
          if gene_symbol!='':
            df=df.append({"Accession":gene,"module":module,"tissue_origin":tissue,"multiple_tissues":"NO","gene_symbol":gene_symbol,"numb_tissues":1,"Sign":sign},ignore_index=True)
      df.to_csv(file_path)
```

```python
## Saving each module as separate file
for tissue in tissues:
    for sign in signs:
        for module in module_dict[(tissue,sign)].keys():
            directory=tissue+'/'+sign+'/'
            df = pd.DataFrame(columns=["Accession","module","tissue_origin","kME_module","multiple_tissues","gene_symbol","entrez_ID","numb_tissues","Sign"])
            genes = module_dict[(tissue,sign)][module]
            for gene in genes:
              gene_symbol=get_gene_symbol(gene)
              df=df.append({"Accession":gene,"module":module,"tissue_origin":tissue,"multiple_tissues":"NO","gene_symbol":gene_symbol,"numb_tissues":1,"Sign":sign},ignore_index=True)
            file_path=directory+module+".csv"
            df.to_csv(file_path)
```
