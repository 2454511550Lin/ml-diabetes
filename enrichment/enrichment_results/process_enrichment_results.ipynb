```python
def get_top_modules(file_path):
  df = pd.read_csv(file_path)
  modules = set(df["module"])
  for module in modules:
    module_data= df.loc[df.module==module]
    max_z_index = df['z.score'].idxmax()
    row_with_z_score = df.loc[max_z_index]
    return row_with_z_score
```

```python
def get_top_modules2(file_path):
  df = pd.read_csv(file_path)
  df= df.loc[df["num.overlap"]>4]
  index = df.groupby('module')['z.score'].idxmax()
  result = df.loc[index]
  return result
```

```python
tissues = ['islet', 'liver', 'adipose', 'kidney', 'gastroc']
signs = ['SIGNED', 'UNSIGNED']
```

```python
df=pd.DataFrame()
for tissue in tissues:
  directory=tissue+'/'
  for sign in signs:
    temp_directory=directory+sign+'/'
    if not os.path.exists(temp_directory):
        print(temp_directory+" doesn't exist.")
    data=pd.DataFrame(get_top_modules2(temp_directory+"enriched_looks_normal.csv"))
    data['SIGN'] = sign
    #print(data)
    if df.empty:
      df=data
    else:
      #print(tissue)
      df=df.append(data, ignore_index=True)
```

```python
df.to_csv("processed_enrichment.csv")
```
